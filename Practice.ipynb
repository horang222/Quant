```python
def find_div_and_EPS(stock, year):
    output = dict()
    try:
        _report = dart.report(stock, "배당", year, '11011')
    except:
        _report = None
        
    if (_report is None) or (len(_report) == 0):
        output['주당배당금'] = np.nan, np.nan, np.nan 
        output['주당순이익'] = np.nan, np.nan, np.nan
        return output
    
    else:
        # 주당 배당금 정리
        div_temp = _report.loc[_report["se"] == "주당 현금배당금(원)"].iloc[0]
        div_row = div_temp[['thstrm', 'frmtrm', 'lwfr']].values
        
        cur_div = int(div_row[0].replace('-','0').replace(',',''))
        pre_div = int(div_row[1].replace('-','0').replace(',',''))
        spre_div = int(div_row[2].replace('-','0').replace(',',''))

        output['주당배당금'] = spre_div, pre_div, cur_div

        # 주당순이익 정리
        EPS_temp = _report.loc[_report["se"].str.contains("주당순이익")].iloc[0]
        EPS_row = EPS_temp[['thstrm', 'frmtrm', 'lwfr']].values

        cur_EPS = int(EPS_row[0].replace('-', '0').replace(',',''))
        pre_EPS = int(EPS_row[1].replace('-', '0').replace(',',''))
        spre_EPS = int(EPS_row[2].replace('-', '0').replace(',',''))
        output['주당순이익']= spre_EPS, pre_EPS, cur_EPS

        return output
```

제출 마감일 계산

```python
import pandas as pd

sub_due_data = [] # 제출 마감일을 담을 데이터

for settle_month in range(1, 13): # 결산 월
    for year in range(2013, 2021): #사업연도
        # 사업연도 경과 후 기준일 계산
        if settle_month < 12:
            after_bs_year_day = pd.to_datetime("{}-{}-1".format(year, settle_month+1))
        else:
            after_bs_year_day = pd.to_datetime("{}-1-1".format(year+1))
        due_date = after_bs_year_day +pd.to_timedelta(90, "D")
        sub_due_data.append([settle_month, year, due_date])
sub_due_data = pd.DataFrame(sub_due_data, columns = ["결산 월", "사업연도", "제출마감일"])
sub_due_data.to_csv("../../데이터/사업보고서_제출마감일.csv", index=False, encoding='euc-kr')
```

장이 열리는 가장 가까운 날짜의 주가 찾기

```python
def find_closest_stock_price(sp_data, date):
    date = pd.to_datetime(date) # 문자열로 입력됐을 가능성이 있는 date를 날짜 자료형으로 변환
    # 주가 데이터를 벗어나는 범위의 날짜가 입력되면 결측을 반환
    if sp_data['Date'].max() < date:
        return np.nan # 주가 데이터의 마지막 날짜보다 입력 날짜가 크다면 원하는 주가가 없는 것이므로 결측을 반환한다. 
    else:
        while True:
            # date와 같은 날짜가 Date에 있으면 
            if sum(sp_data['Date'] == date) > 0: 
                # 해당 날짜의 종가를 저장
                value = sp_data.loc[sp_data['Date'] == date, 'Close'].iloc[0]
                break
            else: # date와 같은 날짜의 Date가 없으면 하루 증가
                date += pd.to_timedelta(1, 'D') 
                # "Date" 컬럼이 date와 같은 요소가 하나라도 없다면 날짜가 date일 때 장이 열리지 않은 것임. 따라서 1일 더해서 다음 날 장이 열렸는지 다시 확인ㅁ
        return value
```

```python
settle_month_dict = stock_list.set_index('Name')['SettleMonth']
settle_month_dict = settle_month_dict.apply(lambda x:int(x[:-1])).to_dict()
sub_due_dict = sub_due_data.set_index(['결산 월', '사업연도'])['제출마감일'].to_dict()

# 라인 1: stock_list에서 'Name' 컬럼을 인덱스로, 'SettleMonth' 컬럼을 데이터로 하는 시리즈를 settle_month_dict에 저장. 
# 라인 2: stock_list의 'SettleMonth' 컬럼에는 '월'이라는 문자열이 포함되어 있음. 이를 문자열 슬라이싱으로 제거하고, 딕셔너리로 변환
# 라인 3: '결산 월'과 '사업연도' 칼럼을 인덱스로, '제출마감일' 컬럼을 데이터로 설정. 

# df[col]과 같이 하나의 컬럼을 지정하면 그 결과는 'series'임에 유의
# 'to_dict()'는 series의 인덱스를 딕셔너리의 키로, 데이터를 딕셔너리의 값으로 변환. 
```
